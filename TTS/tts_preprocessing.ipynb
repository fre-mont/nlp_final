```python
wav_list = os.listdir('./nlp_jjd_dataset/jjd/Training/wavs')
label_list = os.listdir('./nlp_jjd_dataset/jjd/Training/labels')
```

```python
import json
import csv

with open(f'./nlp_jjd_dataset/jjd/Training/labels/{label_list[0]}') as f:
    jsonn = json.load(f)

print(jsonn)

len(jsonn['utterance'])
print(jsonn['utterance'][0]['id'])
print(jsonn['utterance'][0]['start']*1000)
print(jsonn['utterance'][0]['end']*1000)
print(jsonn['utterance'][0]['speaker_id'])
print(jsonn['utterance'][0]["dialect_form"])
```

```python
with open('output.tsv', 'w', newline='', encoding='utf-8') as tsv_file:
    writer = csv.writer(tsv_file, delimiter='\t')
    for wav, label in zip(wav_list, label_list):
        audio = AudioSegment.from_wav(f"./nlp_jjd_dataset/jjd/Training/wavs/{wav}")
        with open(f'./nlp_jjd_dataset/jjd/Training/labels/{label}') as f:
            jsonn = json.load(f)

        for div in range(len(jsonn['utterance'])):
            jsonn['utterance'][div]['id']
            start_time = jsonn['utterance'][div]['start']*1000
            end_time = jsonn['utterance'][div]['end']*1000
            jsonn['utterance'][div]['speaker_id']
            
            writer.writerow([f"{jsonn['utterancxexx'][div]['id']}.wav", jsonn['utterance'][div]["dialect_form"], jsonn['utterance'][div]['speaker_id']])

            cut_audio = audio[start_time:end_time]
```
